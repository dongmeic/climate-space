```python
import os

import pandas as pd
```

```python
DATA_DIR = '../data/'
```

```python
# List data files in DATA_DIR
files = sorted(
    [file for file in os.listdir(DATA_DIR) if file[-4:] == '.csv'])
files
```

```python
filepath = DATA_DIR + files[0]
data = pd.read_csv(filepath)
data.DATE = pd.to_datetime(data.DATE)
data['YEAR'] = data.DATE.apply(lambda x: x.year)
data.head()
```

```python
def split_by_year(filepath, outdir):
    filename = filepath.split('/')[-1]
    data = pd.read_csv(filepath)
    data.DATE = pd.to_datetime(data.DATE)
    yearly_dfs = [data]
    data['YEAR'] = data.DATE.apply(lambda x: x.year)
    years = data.YEAR.unique()
    for year in years:
        yearly_data = data.loc[data.YEAR == year, :]
        outpath = '%s%s_%d.csv' % (outdir, filename, year)
        print('Saving data to %s...' % outpath)
        yearly_data.to_csv(outpath, index=False)
    return True
```

```python
# Test
split_by_year(DATA_DIR + files[0], DATA_DIR + 'clean/')
```
